```python
import pandas as pd
import numpy as np
import datetime
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/Users/elnazshahzeidi/Downloads/learningSet.csv')
data
```

```python
# Check for null values in all columns


null_counts = data.isna().sum()
null_counts
```

```python
# Create a drop list for specified columns

drop_list = ['OSOURCE', 'ZIP']
drop_list
```

```python
# Identify columns with over 85% missing values

missing_percentage = pd.DataFrame(data.isna().sum()*100/len(data), columns=['percentage'])
missing_percentage.sort_values('percentage', ascending = False).head(50)
```

```python
# Remove columns with over 85% missing values

columns_to_drop = list(missing_percentage[missing_percentage['percentage'] >= 85].index)
print(len(columns_to_drop))


drop_list.extend(columns_to_drop)
len(drop_list)
```

```python
# Remove the specified columns from the DataFrame

data.drop(columns=drop_list, inplace=True)
data
```

```python
# Check current value counts in the GENDER column

data['GENDER'].value_counts()
```

```python
# Fill null values
data['GENDER'] = data['GENDER'].fillna('F')  # Assuming 'F' for nulls; adjust if necessary

# Use a lambda function to categorize the GENDER values
data['GENDER'] = data['GENDER'].apply(lambda x: 'M' if x in ['M', 'Male'] else ('F' if x in ['F', 'Female'] else 'Other'))

```
